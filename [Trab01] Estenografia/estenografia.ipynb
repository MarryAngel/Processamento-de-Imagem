```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mostrar_imagem(imagem, titulo):
    plt.imshow(imagem)
    plt.title(titulo)
    plt.show()
```

```python
img = cv2.imread("baboon.png")
imagem = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
plt.title('my picture')
plt.show()
```

```python
(canalVermelho, canalVerde, canalAzul) = cv2.split(imagem) 
mostrar_imagem(canalVermelho, "Canal Vermelho")
mostrar_imagem(canalVerde, "Canal Verde")
mostrar_imagem(canalAzul, "Canal Azul")
```

```python
resultado = cv2.merge([canalVermelho, canalVerde, canalAzul]) 
mostrar_imagem(resultado, "Imagem Original")
```

```python
zeros = np.zeros(imagem.shape[:2], dtype = "uint8")
img_vermelho = cv2.merge([canalVermelho,zeros, zeros])
img_verde = cv2.merge([zeros, canalVerde, zeros])
img_azul = cv2.merge([zeros, zeros, canalAzul])
mostrar_imagem(img_vermelho, "Canal Vermelho")
mostrar_imagem(img_verde, "Canal Verde")
mostrar_imagem(img_azul, "Canal Azul")
```

```python
# primeiro pixel do canal vermelho
pixel = canalVermelho[0, 0]
print(pixel)
# converter pixel para binário
pixel_binario = bin(pixel)
print(pixel_binario)

# pegar o bit 0, 1, 2 e 7
bit0 = pixel_binario[-1]
bit1 = pixel_binario[-2]
bit2 = pixel_binario[-3]
bit7 = pixel_binario[-8]
print(bit7, bit2, bit1, bit0)
```

```python
# converter todos os pixels do canal vermelho para binário e colocálos numa lista
pixels_binario = [bin(pixel) for pixel in canalVermelho.flatten()]

# pegar o bit 0 de todos os pixels
bits0 = [pixel[-1] for pixel in pixels_binario]

# mostrar como imagem binária em preto e branco
imagem_binaria = np.array(bits0, dtype='uint8')
imagem_binaria = imagem_binaria.reshape(canalVermelho.shape)
mostrar_imagem(imagem_binaria, "Bit 0")
```
